```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import os
import cv2 as cv
import random
import time
import psutil
import sqlite3 
from datetime import datetime, timedelta
```

```python
NUM_CAMERAS = 1
FPS = 5  
NUM_FRAMES = 3600  
model = keras.models.load_model('model.h5')

conn = sqlite3.connect('presence_log.db')
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS presence_log (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        camera_id INTEGER,
        entry_time TEXT,
        exit_time TEXT,
        frame_count INTEGER
    )
''')

presence = [0] * NUM_CAMERAS
entry_time = [None] * NUM_CAMERAS
frame_counter = [0] * NUM_CAMERAS
time_on = [0] * NUM_CAMERAS
time_off = [0] * NUM_CAMERAS

cap = cv.VideoCapture('Video_2025-06-13_060505.wmv')
count = 0
i = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("End of video stream or error reading frame.")
        break

    img = frame[300:600,200:800]
    img = cv.cvtColor(img,  cv.COLOR_RGB2GRAY) / 255
    img = cv.resize(img, (128, 128))
    img = np.expand_dims(img, axis=0)
    img = np.expand_dims(img, axis=-1)
    #cv.imwrite(os.path.join('/content/images', f"{count}.jpg"), img)
    preds = model.predict(img)
    labels = (preds >= 0.5).astype(int)

    for j in range(NUM_CAMERAS):
        detected = labels[j][0]

        if presence[j] == 0 and detected == 1:
            presence[j] = 1
            entry_time[j] = datetime.now()
            frame_counter[j] = 1
            time_on[j] += 1

        elif presence[j] == 1 and detected == 1:
            frame_counter[j] += 1
            time_on[j] += 1

        elif presence[j] == 1 and detected == 0:
            presence[j] = 0
            exit_time = entry_time[j] + timedelta(seconds=frame_counter[j] / FPS)

            cursor.execute('''
                INSERT INTO presence_log (camera_id, entry_time, exit_time, frame_count)
                VALUES (?, ?, ?, ?)
            ''', (
                j,
                entry_time[j].strftime('%Y-%m-%d %H:%M:%S'),
                exit_time.strftime('%Y-%m-%d %H:%M:%S'),
                frame_counter[j]
            ))
            conn.commit()
            frame_counter[j] = 0
            time_off[j] += 1
        else:
            time_off[j] += 1

        
        on_seconds = time_on[j] // FPS
        off_seconds = time_off[j] // FPS

        on_str = f"ON: {on_seconds//3600:02}:{(on_seconds%3600)//60:02}:{on_seconds%60:02}"
        off_str = f"OFF: {off_seconds//3600:02}:{(off_seconds%3600)//60:02}:{off_seconds%60:02}"
        total = time_on[j] + time_off[j]
        p_on = (time_on[j] / total) * 100 if total > 0 else 0
        p_off = 100 - p_on
        percent_on = f"({p_on:.1f})%"
        percent_off = f"({p_off:.1f})%"

        start_point = (200, 300)  
        end_point = (800, 600)  
        color = (0, 255, 0)      
        thickness = 3         
        
        frame = cv.rectangle(frame, start_point, end_point, color, thickness)
        cv.putText(frame, on_str, (200, 250), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0,0), 3)
        cv.putText(frame, off_str, (200, 290), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 3)
        cv.putText(frame, percent_on, (420, 250), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 3)
        cv.putText(frame, percent_off, (420, 290), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 3)

    

    cv.imwrite(os.path.join('pred_4', f"{i}.png"), frame)
    i += 1
    
conn.close()
```
