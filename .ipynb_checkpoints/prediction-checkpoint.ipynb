```python
import os
import cv2 as cv
import numpy as np
import sqlite3
from datetime import datetime, timedelta

NUM_CAMERAS = 6
FPS = 5  
NUM_FRAMES = 600  
model = keras.models.load_model('model.h5')

conn = sqlite3.connect('presence_log.db')
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS presence_log (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        camera_id INTEGER,
        entry_time TEXT,
        exit_time TEXT,
        frame_count INTEGER
    )
''')

presence = [0] * NUM_CAMERAS
entry_time = [None] * NUM_CAMERAS
frame_counter = [0] * NUM_CAMERAS
time_on = [0] * NUM_CAMERAS
time_off = [0] * NUM_CAMERAS

os.makedirs("pred_2", exist_ok=True)

for i in range(NUM_FRAMES):
    gray_images = [
    cv.resize(cv.cvtColor(images[j][i].astype(np.uint8), cv.COLOR_RGB2GRAY), (128, 128))
    for j in range(NUM_CAMERAS)
]

    
    img_batch = np.stack(gray_images) / 255.0
    img_batch = np.expand_dims(img_batch, axis=-1)  # Shape: (6, 128, 128, 1)

    preds = model.predict(img_batch)
    labels = (preds >= 0.5).astype(int)

    for j in range(NUM_CAMERAS):
        detected = labels[j][0]

        if presence[j] == 0 and detected == 1:
            presence[j] = 1
            entry_time[j] = datetime.now()
            frame_counter[j] = 1
            time_on[j] += 1

        elif presence[j] == 1 and detected == 1:
            frame_counter[j] += 1
            time_on[j] += 1

        elif presence[j] == 1 and detected == 0:
            presence[j] = 0
            exit_time = entry_time[j] + timedelta(seconds=frame_counter[j] / FPS)

            cursor.execute('''
                INSERT INTO presence_log (camera_id, entry_time, exit_time, frame_count)
                VALUES (?, ?, ?, ?)
            ''', (
                j,
                entry_time[j].strftime('%Y-%m-%d %H:%M:%S'),
                exit_time.strftime('%Y-%m-%d %H:%M:%S'),
                frame_counter[j]
            ))
            conn.commit()
            frame_counter[j] = 0
            time_off[j] += 1
        else:
            time_off[j] += 1

        frame = images[j][i]
        on_seconds = time_on[j] // FPS
        off_seconds = time_off[j] // FPS

        on_str = f"ON: {on_seconds//3600:02}:{(on_seconds%3600)//60:02}:{on_seconds%60:02}"
        off_str = f"OFF: {off_seconds//3600:02}:{(off_seconds%3600)//60:02}:{off_seconds%60:02}"
        total = time_on[j] + time_off[j]
        p_on = (time_on[j] / total) * 100 if total > 0 else 0
        p_off = 100 - p_on
        percent_on = f"({p_on:.1f})%"
        percent_off = f"({p_off:.1f})%"

        cv.putText(frame, on_str, (10, 20), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        cv.putText(frame, off_str, (10, 40), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        cv.putText(frame, percent_on, (120, 20), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        cv.putText(frame, percent_off, (120, 40), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    row1 = np.hstack((images[0][i], images[1][i], images[2][i]))
    row2 = np.hstack((images[3][i], images[4][i], images[5][i]))
    combined = np.vstack((row1, row2))

    cv.imwrite(os.path.join('pred_4', f"{i}.png"), combined)

conn.close()
```
